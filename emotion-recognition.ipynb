```python
img = cv2.imread('angry.jpeg')
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
from deepface import DeepFace
```

```python
predictions = DeepFace.analyze(img, actions=("emotion"))
```

```python
#type(predictions)
```

```python
# predictions
```

```python
emotion = predictions[0]['dominant_emotion']
```

```python
face_cascade = cv2.CascadeClassifier('haar_face.xml')
```

```python
grayimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# plt.imshow(grayimg)
```

```python
faces = face_cascade.detectMultiScale(grayimg, 1.1, 3)
```

```python
# faces
```

```python
for a,b,c,d in faces:
    cv2.rectangle(img, (a,b), (a+c,b+d), (255,0,0), 3)
```

```python
font_scale = 2.0  
text_size = cv2.getTextSize(emotion, cv2.FONT_HERSHEY_SIMPLEX, font_scale, 2)[0]
text_position = (a + c // 2 - text_size[0] // 2, b + d + 50)  

cv2.rectangle(img, (text_position[0], text_position[1] - text_size[1]), 
              (text_position[0] + text_size[0], text_position[1]), (0, 0, 0), -1)
cv2.putText(img, emotion, text_position, cv2.FONT_HERSHEY_SIMPLEX, font_scale, (255, 255, 255), 2)


plt.imshow(img)
```
